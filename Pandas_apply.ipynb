```python
import pandas as pd
```

```python
movie_rating = pd.read_csv('ratings.csv')
movie_rating.head()
```

```python
def classify_movie(row):
    ''' Классифицирует фильмы по рейтингу'''
    if row['rating'] == 5 or row['rating'] == 4.5:
        return 'Высокий рейтинг'
    elif row['rating'] <= 4:
        return 'Средний рейтинг'
    elif row['rating'] <= 2:
        return 'Низкий рейтинг' 
```

```python
movie_rating['class'] = movie_rating.apply(classify_movie, axis = 1)
movie_rating.head(20)
```

```python
keywords = pd.read_csv('keywords.csv')
keywords.head(10)
```

```python
def classif_region (row):
    '''Геоклассификатор в соответствии
    geo_data = {

    'Центр': ['москва', 'тула', 'ярославль'],

    'Северо-Запад': ['петербург', 'псков', 'мурманск'],

    'Дальний Восток': ['владивосток', 'сахалин', 'хабаровск']
    }
    '''
    geo_data = {

    'Центр': ['москва', 'тула', 'ярославль'],

    'Северо-Запад': ['петербург', 'псков', 'мурманск'],

    'Дальний Восток': ['владивосток', 'сахалин', 'хабаровск']
    }
    for region, keyword_ in geo_data.items(): 
        for word in keyword_:            
            if word in row['keyword']:
                return region
    
    return 'undefined'
        
```

```python
keywords['region'] = keywords.apply(classif_region, axis = 1)
keywords.head(5)
```

```python
year = [str(n) for n in list(range(1950,2011))]
print(year)
```

```python
movies_names = pd.read_csv('movies.csv')
movies_names.head()
```

```python
def production_year(row):
    year_from_title = row['title'][-5:-1]
#     year_n = year_from_title
    if year_from_title in year:
        return year_from_title
    else:
        return '1900'
```

```python
movies_names['year'] = movies_names.apply(production_year, axis = 1)
movies_full = movies_names.merge(movie_rating, on = 'movieId', how = 'left')
avg_rating = movie_full.groupby('year').mean().reset_index()[['year', 'rating']].sort_values(by = 'rating', ascending = False)
avg_rating.head(20)
```
